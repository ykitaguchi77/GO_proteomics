```python
from google.colab import drive
drive.mount("/content/drive")

import pandas as pd
import numpy as np

pd.set_option('display.max_rows', 200)
pd.set_option('display.max_columns', 200)

```

```python
# Reload the CSV files into DataFrames using the Shift-JIS encoding
df_disease_info = pd.read_csv('/content/drive/MyDrive/研究/進行中の研究/甲状腺眼症プロテオミクス/甲状腺眼症採血/●DiseaseInfo_modified1.5_connected_9800.csv', encoding='shift-jis')
df_patient_labo = pd.read_csv('/content/drive/MyDrive/研究/進行中の研究/甲状腺眼症プロテオミクス/甲状腺眼症採血/patient_labodata_list.csv', encoding='cp932')

# Convert the 'PATIENTNO' in df_patient_labo and 'Unnamed: 0' in df_disease_info to integer type
df_patient_labo['PATIENTNO'] = df_patient_labo['PATIENTNO'].astype(int)
df_disease_info['Unnamed: 0'] = df_disease_info['Unnamed: 0'].astype('Int64', errors='ignore')

# Find matching rows based on 'PATIENTNO' and 'Unnamed: 0'
matching_rows = pd.merge(df_patient_labo, df_disease_info, left_on='PATIENTNO', right_on='Unnamed: 0', how='inner')



# Create new columns in df_patient_labo and initialize them with None
new_col_names_8th = 'severity'
new_col_names_9th = 'activity'
df_patient_labo[new_col_names_8th] = None
df_patient_labo[new_col_names_9th] = None

# Create a mapping dictionary from 'PATIENTNO' to 8th and 9th columns in matching_rows
col_name_8th = matching_rows.columns[44]
col_name_9th = matching_rows.columns[45]
map_dict_8th = dict(zip(matching_rows['PATIENTNO'], matching_rows[col_name_8th]))
map_dict_9th = dict(zip(matching_rows['PATIENTNO'], matching_rows[col_name_9th]))

# Use the mapping dictionary to populate the new columns in df_patient_labo
df_patient_labo[new_col_names_8th] = df_patient_labo['PATIENTNO'].map(map_dict_8th)
df_patient_labo[new_col_names_9th] = df_patient_labo['PATIENTNO'].map(map_dict_9th)

# 患者IDの匿名化
# 一意なPATIENTNOを取得
unique_patient_no = df_patient_labo['PATIENTNO'].unique()
# 一意なPATIENTNOに通し番号を割り当てる
mapping = {patient_no: i for i, patient_no in enumerate(unique_patient_no)}
# PATIENTNO列を通し番号に置き換え
df_patient_labo['PATIENTNO'] = df_patient_labo['PATIENTNO'].map(mapping)


# 患者氏名の列を削除
df_patient_labo = df_patient_labo.drop("漢字氏名", axis=1)

df_patient_labo
```

```python
#############################
# 各患者に関して初回の採血のみを抜粋
#############################

# '検査日'列を日付型に変換
df_patient_labo['検査日'] = pd.to_datetime(df_patient_labo['検査日'], errors='coerce')

# 同じPATIENTNOに対して、検査日が古いもののみ残す
df_patient_labo_filtered = df_patient_labo.sort_values(by=['PATIENTNO', '検査日']).drop_duplicates(subset='PATIENTNO', keep='first')

df_patient_labo_filtered
```

```python
df = df_patient_labo_filtered
df_active = df[df["activity"]=="活動期（ステロイドあり）"]
df_others = df[df["activity"]!="活動期（ステロイドあり）"]
print(f"Active GO: {len(df_active)}, No or inactive GO: {len(df_others)}")
```

```python
df_active.to_csv("/content/drive/MyDrive/研究/進行中の研究/甲状腺眼症プロテオミクス/甲状腺眼症採血/activeGO.csv", encoding='cp932', index=False)
df_others.to_csv("/content/drive/MyDrive/研究/進行中の研究/甲状腺眼症プロテオミクス/甲状腺眼症採血/othersGO.csv", encoding='cp932', index=False)
```
