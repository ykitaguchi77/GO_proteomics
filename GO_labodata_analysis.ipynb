```python
# Reload the CSV files into DataFrames using the Shift-JIS encoding
df_disease_info = pd.read_csv('/content/drive/MyDrive/研究/進行中の研究/甲状腺眼症プロテオミクス/甲状腺眼症採血/●DiseaseInfo_modified1.5_connected_9800.csv', encoding='shift-jis')
df_patient_labo = pd.read_csv('/content/drive/MyDrive/研究/進行中の研究/甲状腺眼症プロテオミクス/甲状腺眼症採血/patient_labodata_list.csv', encoding='cp932')

# Convert the 'PATIENTNO' in df_patient_labo and 'Unnamed: 0' in df_disease_info to integer type
df_patient_labo['PATIENTNO'] = df_patient_labo['PATIENTNO'].astype(int)
df_disease_info['Unnamed: 0'] = df_disease_info['Unnamed: 0'].astype('Int64', errors='ignore')

# Find matching rows based on 'PATIENTNO' and 'Unnamed: 0'
matching_rows = pd.merge(df_patient_labo, df_disease_info, left_on='PATIENTNO', right_on='Unnamed: 0', how='inner')



# Create new columns in df_patient_labo and initialize them with None
new_col_names_8th = 'severity'
new_col_names_9th = 'activity'
df_patient_labo[new_col_names_8th] = None
df_patient_labo[new_col_names_9th] = None

# Create a mapping dictionary from 'PATIENTNO' to 8th and 9th columns in matching_rows
col_name_8th = matching_rows.columns[44]
col_name_9th = matching_rows.columns[45]
map_dict_8th = dict(zip(matching_rows['PATIENTNO'], matching_rows[col_name_8th]))
map_dict_9th = dict(zip(matching_rows['PATIENTNO'], matching_rows[col_name_9th]))

# Use the mapping dictionary to populate the new columns in df_patient_labo
df_patient_labo[new_col_names_8th] = df_patient_labo['PATIENTNO'].map(map_dict_8th)
df_patient_labo[new_col_names_9th] = df_patient_labo['PATIENTNO'].map(map_dict_9th)

# 患者IDの匿名化
# 一意なPATIENTNOを取得
unique_patient_no = df_patient_labo['PATIENTNO'].unique()
# 一意なPATIENTNOに通し番号を割り当てる
mapping = {patient_no: i for i, patient_no in enumerate(unique_patient_no)}
# PATIENTNO列を通し番号に置き換え
df_patient_labo['PATIENTNO'] = df_patient_labo['PATIENTNO'].map(mapping)


# 患者氏名の列を削除
df_patient_labo = df_patient_labo.drop("漢字氏名", axis=1)

df_patient_labo
```

```python
#############################
# 各患者に関して初回の採血のみを抜粋
#############################

# '検査日'列を日付型に変換
df_patient_labo['検査日'] = pd.to_datetime(df_patient_labo['検査日'], errors='coerce')

# 同じPATIENTNOに対して、検査日が古いもののみ残す
df_patient_labo_filtered = df_patient_labo.sort_values(by=['PATIENTNO', '検査日']).drop_duplicates(subset='PATIENTNO', keep='first')

df_patient_labo_filtered
```

```python
df = df_patient_labo_filtered

# Assuming df is your DataFrame
df_active = pd.concat([
    df[df["activity"] == "活動期（ステロイドあり）"],
    df[df["activity"] == "活動期（ステロイドなし）"]
])
df_inactive = df[df["activity"]=="不活動期"]
df_others = df[~df['activity'].isin(['活動期（ステロイドあり）', '活動期（ステロイドなし）'])]

print(f"Active GO: {len(df_active)}, inactive GO: {len(df_inactive)}, others: {len(df_others)}")
```

```python
"""
データのクリーンニング
"""
# Columns to be compared
columns_to_compare = [
    "WBC(白血球数)", "RBC(赤血球数)", "Hb(ﾍﾓｸﾞﾛﾋﾞﾝ)", "Ht(ﾍﾏﾄｸﾘｯﾄ)", "ＭＣＶ", "ＭＣＨ", "ＭＣＨＣ",
    "好中球(機械法)", "ﾘﾝﾊﾟ球(機械法)", "好酸球(機械法)", "AST(GOT)", "ALT(GPT)", "T-Cho(総ｺﾚｽﾃﾛｰﾙ)",
    "TG(中性脂肪)", "ＣＲＰ", "TP(総ﾀﾝﾊﾟｸ)", "ＦＴ３", "ＴＲＡｂ", "ＴＳＡｂ", "TPOAb", "ＴＳＨ", "ＩｇＧ"
]

# Function to clean and convert the data with additional handling for '>', '<', and 'ﾐﾏﾝ'
def clean_and_convert_final(df, columns):
    for col in columns:
        # Replace '<', '>', and 'ﾐﾏﾝ' with empty string and convert to float
        df[col] = df[col].apply(lambda x: np.nan if pd.isna(x) else float(str(x).replace('<', '').replace('>', '').replace('ﾐﾏﾝ', '')))
    return df

# Clean and convert the data in both DataFrames
df_active_clean = clean_and_convert_final(df_active.copy(), columns_to_compare)
df_inactive_clean = clean_and_convert_final(df_inactive.copy(), columns_to_compare)
df_others_clean = clean_and_convert_final(df_others.copy(), columns_to_compare)

# Show the first few rows of the cleaned DataFrames for verification
df_active_clean.head()
```

```python
"""
Compare labodata between active and others
"""

# Initialize the comparison results as an empty dictionary
comparison_results = {}

# Loop through each column to compare
for col in columns_to_compare:
    # Skip the column if it doesn't exist in either DataFrame
    if col not in df_active_clean.columns or col not in df_others_clean.columns:
        continue

    # Extract the data for the column from both DataFrames
    active_data = df_active_clean[col].dropna()
    inactive_data = df_inactive_clean[col].dropna()
    others_data = df_others_clean[col].dropna()

    # Create a subplot for the comparison
    plt.figure(figsize=(12, 6))
    sns.histplot(active_data, kde=True, color='b', label='Active', bins=50, alpha=0.5)
    sns.histplot(inactive_data, kde=True, color='b', label='Active', bins=50, alpha=0.5)
    sns.histplot(others_data, kde=True, color='r', label='Others', bins=50, alpha=0.5)
    plt.title(f'Comparison of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()

    # Compute summary statistics for the column for both DataFrames
    summary_active = active_data.describe()
    summary_inactive = inactive_data.describe()
    summary_others = others_data.describe()

    # Store the summary statistics in the comparison results dictionary
    comparison_results[col] = {'active': summary_active, 'inactive': summary_inactive,'others': summary_others}

# Show some of the comparison results as an example
#example_keys = list(comparison_results.keys())[:3]
example_keys = list(comparison_results.keys())
for key in example_keys:
    print(f"{key}: {comparison_results[key]}\n")
```

```python
from scipy.stats import ttest_ind

# Initialize a dictionary to store the p-values for each column
p_values = {}

# Loop through each column to perform t-tests
for col in columns_to_compare:
    # Skip the column if it doesn't exist in either DataFrame
    if col not in df_active_clean.columns or col not in df_others_clean.columns:
        continue

    # Extract the data for the column from both DataFrames
    active_data = df_active_clean[col].dropna()
    others_data = df_inactive_clean[col].dropna()

    # Perform an independent t-test
    t_stat, p_val = ttest_ind(active_data, others_data, equal_var=False, nan_policy='omit')

    # Store the p-value in the dictionary
    p_values[col] = p_val

# Show the p-values
p_values
```
